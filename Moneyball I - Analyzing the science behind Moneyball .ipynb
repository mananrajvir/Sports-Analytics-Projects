* **Basis of Moneyball**:
    * Before Billy Beane, in baseball most of the scouts and teams measured a batter and the hitting ability by statistics like Slugging Percentage (how hard you can hit). None of these teams valued the player's ability to draw a walk (after 4 balls).
    * Billy Beane focussed more on on-base percentage, that took into consideration the players ability to walk, since ideally it doesn't make a difference if the player hits or walks to first base. Since drawing a walk wasn't considered an important skill by other players - these player's were very cheap to buy!

* **Slugging Percentage(SLG):**
    * **SLG = (Singles + 2 * Doubles + 3 * Triples + 4 * HomeRuns) / At bats**
    * Single = 1b. Double = 2b. Triple = 3b. HomeRun = hr. At bats = ab.
    * Measure of the batting productivity of a hitter.
    * SLG = Total bases (number of bases a player has gained with hits) / At bats (batters turn batting against a pitcher)
    * Unlike on-base percentage, slugging percentage deals only with hits and **does not include walks** and hit-by-pitches in its equation.
    * Slugging percentage differs from batting average in that all hits are not valued equally.


* **On-Base Percentage:**
    * **OBP = (Hits + Walks + Hit by Pitch) / (All bats + Hit by Pitch + Sacrifice Flies)**
    * hit = h. walk = bb. Hit by Pitch = hbp. Sacrifice Flies = sf.
    * Sacrifice Plies = when a batter hits the ball and is caught, but in the process a runner on base scores a run.

## Analyzing the Impact of On Base and Slugging Percentage on Winning

* Uses **regression analysis:**
    * mechanism for establishing the relationship between a variable and another group of variables.


* Show that on base percentage is more statistically significant in determing success, as compared to slugging percentage.


* **Challenges & Solutions:**
    * The success rate / win rate, will depend not just on the on-base & slugging percentage of a team, but also on the performance of their opponent
        * Sol - First calculate the on-base & slugging percentage of each team, across the season. Then work out the on-base and slugging percentage statistics against each team (success of opponents agains the team, across the season). Run regressions on this wherein the target variable is the win percentage.
        * Reg1 --> Looks at onbase percentage and onbase percentage against for each team.
        * Reg2 --> Looks at the slugging percentage and slugging percentage aginst for each team.
        * Reg3 --> Include both onbase and slugging percentage, for and against for each team.
        * Reg4 --> Same as 3 but imposes that the the magnitude should be equal for - for & against of each percentage.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
Teams = pd.read_excel("datasets/Game logs 1999-2003.xlsx")
```

```python
print(Teams.columns.tolist())
```

```python
"""Create a variable to identify the winner of the game [using home_score and visitor_score]"""
Teams["hwin"] = np.where(Teams["home_score"] > Teams["visitor_score"], 1, 0)
Teams["awin"] = np.where(Teams["home_score"] < Teams["visitor_score"], 1, 0)
Teams["tie"] = np.where(Teams["home_score"] == Teams["visitor_score"], 1, 0)
```

```python
"""Create a variable that defines the season for each game [extract from the date of the game]""" 
Teams["year"] = Teams["date"].astype(str).str[0:4]  #date format is: YYYYMMDD --> change to string and get 1st 4 characters
```

```python
display(Teams)
```

**Aggregate the required statistics:**
* ab - at bat
* h - hits
* 1b - singles [= h - 2b - 3b - hr]
* 2b - double (second base)
* 3b - triple (third base)
* hr - homerun
* sf - sacrifice flies
* bb - walk
* hbp - hit by pitch

### Create 2 databases:
* For our analysis we need both: slugging & on-base percentage of a team across a season AND slugging & on-base percentage against that team (of the opponents) across the season. 
* To aggeregate the data we need to generate all of the home and away records for each team. So we will create 2 aggregations and then merge them to create a single set of records for each team in the season.

```python
# Measure & Aggregate performance of each team while playing at home
Teamshome = Teams.groupby(["home", "year"])["visitor_ab", "visitor_h", "visitor_2b", "visitor_3b", "visitor_hr", 
                                              "visitor_sf", "visitor_hbp", "visitor_bb", "home_ab", "home_h", "home_2b", 
                                              "home_3b", "home_hr", "home_sf", "home_hbp", "home_bb", "hwin", "awin"].sum()
Teamshome = Teamshome.reset_index()
Teamshome = Teamshome.rename(columns = {"home":"Team"})
Teamshome
```

### For home team, you need to group first by team name, and then by the season. This is because the team will play several home games every year

```python
# Measure & Aggregate performance of each team while playing away
Teamsaway = Teams.groupby(["year", "visitor"])["visitor_ab", "visitor_h", "visitor_2b", "visitor_3b", "visitor_hr", 
                                              "visitor_sf", "visitor_hbp", "visitor_bb", "home_ab", "home_h", "home_2b", 
                                              "home_3b", "home_hr", "home_sf", "home_hbp", "home_bb", "hwin", "awin"].sum()
Teamsaway = Teamsaway.reset_index()
Teamsaway = Teamsaway.rename(columns = {"visitor":"Team"})
Teamsaway
```

### For away teams, we group first by year and then by the team.

```python
"""Merge the 2 datasets"""
Teams2 = pd.merge(Teamshome, Teamsaway, on = ["Team", "year"])
pd.set_option('display.max_columns', 50)
display(Teams2)
```

### Merged Dataset
* The dataframe contains the records for each team, in each season.
* All the '_x'  variables relate to the team as home team
* All the '_y' variables relate to the team as away team


* For Team ANA:
    * visitor_ab_x = at bats of the opposite team when ANA played at home
    * home_ab_x = at bats of ANA when they played at home
    * visitor_ab_y = at bats of ANA (visitor team) when ANA played away
    * home_ab_y = at bats of opposite team (home team) when ANA played away 

```python
"""Total number of wins for each team in each season"""
Teams2["wins"] = Teams2["hwin_x"] + Teams2["awin_y"] #home wins when playing as hometeam + away wins when playing as away team
```

### Calculate the slugging percentage and on-base percentage

```python
"""On-base percentage [OBP] = (h + bb + hbp) / (ab + bb + hpb + sf) for and against each team"""

#First calculate OBP for a team [OBPFOR]: OBP of team while playing at home + OBP of team while playing as away team
Teams2["OBPFOR"] = ((Teams2['home_h_x'] + Teams2['visitor_h_y'] + Teams2['home_bb_x'] + Teams2['visitor_bb_y']
                    + Teams2['home_hbp_x'] + Teams2['visitor_hbp_y']) / (Teams2['home_ab_x'] + Teams2['visitor_ab_y']
                    + Teams2['home_bb_x'] + Teams2['visitor_bb_y'] + Teams2['home_hbp_x'] + Teams2['visitor_hbp_y']
                    + Teams2['home_sf_x'] + Teams2['visitor_sf_y']))
```

```python
#Calculate OBP against a team [OBPAGN]: OBP against a team while they play at home + OBP against the team when playing away
Teams2["OBPAGN"] = ((Teams2['home_h_y'] + Teams2['visitor_h_x'] + Teams2['home_bb_y'] + Teams2['visitor_bb_x']
                    + Teams2['home_hbp_y'] + Teams2['visitor_hbp_x']) / (Teams2['home_ab_y'] + Teams2['visitor_ab_x']
                    + Teams2['home_bb_y'] + Teams2['visitor_bb_x'] + Teams2['home_hbp_y'] + Teams2['visitor_hbp_x']
                    + Teams2['home_sf_y'] + Teams2['visitor_sf_x']))
```

```python
"""Slogging percentage [SLG] = ((h - 2b - 3b - hr) + 2(2b) + 3(3b) + 4(hr)) / ab"""

#First calculate SLG for a team [SLGFOR]: SLG of team while playing at home + SLG of team while playing as away team
Teams2["SLGFOR"] = ((Teams2["home_h_x"] + Teams2["visitor_h_y"] - 
                     (Teams2["home_2b_x"] + Teams2["visitor_2b_y"]) - 
                     (Teams2["home_3b_x"] + Teams2["visitor_3b_y"]) -
                     (Teams2["home_hr_x"] + Teams2["visitor_hr_y"]) + 
                     2*Teams2["home_2b_x"] + 2*Teams2["visitor_2b_y"] + 
                     3*Teams2["home_3b_x"] + 3*Teams2["visitor_3b_y"] +
                     4*Teams2["home_hr_x"] + 4*Teams2["visitor_hr_y"]) /
                     (Teams2["home_ab_x"] + Teams2["visitor_ab_y"]))
```

```python
#Calculate SLG against a team [SLGAGN]: SLG against a team while they play at home + SLG against the team when playing away
Teams2["SLGAGN"] = ((Teams2["home_h_y"] + Teams2["visitor_h_x"] - 
                     (Teams2["home_2b_y"] + Teams2["visitor_2b_x"]) - 
                     (Teams2["home_3b_y"] + Teams2["visitor_3b_x"]) -
                     (Teams2["home_hr_y"] + Teams2["visitor_hr_x"]) + 
                     2*Teams2["home_2b_y"] + 2*Teams2["visitor_2b_x"] + 
                     3*Teams2["home_3b_y"] + 3*Teams2["visitor_3b_x"] +
                     4*Teams2["home_hr_y"] + 4*Teams2["visitor_hr_x"]) /
                     (Teams2["home_ab_y"] + Teams2["visitor_ab_x"]))
```

```python
"""Create a new dataset, with only the required columns"""
Teams3 = Teams2[["year", "Team", "wins", "OBPFOR", "OBPAGN", "SLGFOR", "SLGAGN"]]
Teams3
```

### Calculate the win percentage for each team
* For win percentage, we need the number of games played by each team: use groupby on Teams df on year and home --> .count()

```python
TeamsGh = Teams.groupby(['year', 'home'])['hwin'].count()   #home games played by each team
TeamsGh = TeamsGh.reset_index()
TeamsGh = TeamsGh.rename(columns = {"home":"Team"})

TeamsGa = Teams.groupby(['year', 'visitor'])['awin'].count()  #away games played by each team
TeamsGa = TeamsGa.reset_index()
TeamsGa = TeamsGa.rename(columns = {"visitor":"Team"})

TeamsG = pd.merge(TeamsGh, TeamsGa, on = ["Team", "year"])
TeamsG["Games"] = TeamsG["hwin"] + TeamsG["awin"]
TeamsG = TeamsG[["year", "Team", "Games"]]  #total games played by each team
TeamsG
```

```python
Teams3 = pd.merge(Teams3, TeamsG, on = ["Team", "year"])
Teams3.head(10)
```

```python
"""Calculate Win Percentage"""
Teams3["wpc"] = Teams3["wins"] / Teams3["Games"]
Teams3
```

### Running the Regression

```python
import statsmodels.formula.api as smf 
```

```python
winOBP_lm = smf.ols(formula = "wpc ~ OBPFOR + OBPAGN", data = Teams3).fit()
winOBP_lm.summary()
```

```python
winSLG_lm = smf.ols(formula = "wpc ~ SLGFOR + SLGAGN", data = Teams3).fit()
winSLG_lm.summary()
```

```python
winOBPSLG_lm = smf.ols(formula = "wpc ~ OBPFOR + OBPAGN + SLGFOR + SLGAGN", data = Teams3).fit()
winOBPSLG_lm.summary()
```

```python
winOBPSLGR_lm = smf.ols(formula = "wpc ~ I(OBPFOR-OBPAGN) + I(SLGFOR-SLGAGN)", data = Teams3).fit()
winOBPSLGR_lm.summary()
```

```python
"""Combine all the results together"""
from statsmodels.iolib.summary2 import summary_col
Header = ["1","2","3","4"]
Table_1 = summary_col([winOBP_lm, winSLG_lm, winOBPSLG_lm, winOBPSLGR_lm], 
                      regressor_order = ['Intercept', 'OBPFOR', 'OBPAGN', 'SLGFOR', 'SLGAGN'], model_names = Header)
print(Table_1)
```

#### Interpretations of the result:
* The coefficient value of OBPFOR (3.3009) is greater than that of SLGFOR (1.7552) --> **The onbase percentage, is more important than the slugging percentage, when determing the win percentage of a team**
* This result therefore shows that ideally, the teams should have given a significant consideration to OBP when determing a player's salary!
